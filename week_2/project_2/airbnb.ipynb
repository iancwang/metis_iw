```python
from __future__ import print_function, division
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import re
```

```python
chromedriver = "/Users/iancwang/Downloads/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)

url = 'https://www.airbnb.com/rooms/15776053?location=Seattle%2C%20WA%2C%20United%20States&s=kN_OgJN3'

driver.get(url)
```

```python
amenities = driver.find_element_by_xpath('//*[@id="room"]/div/div[3]/div[1]/div/div[2]/div[1]/div/div/div[4]/div/div/div/section/div[3]/div/button/span')
amenities.click()

soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
#parse listing ID in URL to use as html file name

urlid= re.findall('[/]\d*', url)
urlid= urlid[3]
urlid=urlid.strip('/')
```

```python
#Create file for amenities:

f = open('amenities{}.txt'.format(urlid),'w')
f.write(soup.text)
f.close()
```
