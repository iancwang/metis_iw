```python
# Start cleaning the data

tech_data = tech_data.replace(' ','')
tech_data = tech_data.replace('\n\n','\n')
```

```python
# Withdraw useful info - name, number of employees, and zip code and make a dictionary

company_dict = {}

end = 0
    
for i in range(50):  
    # Find the names
    start_name = tech_data.find('\n', end)
    end_name = tech_data.find('\n',start_name+1)
    name = tech_data[start_name+1:end_name]
    
    company = name[0:len(name)-5]
    zip_code = int(name[-5:])

    #Find the number of employees
    start_emp = tech_data.find('|',end_name)
    end_emp = tech_data.find('|',start_emp+1)
    emp = int((tech_data[start_emp+11:end_emp]).replace(',',''))

    company_dict[company] = [emp, zip_code]
    
    end = tech_data.find('\n\n', end_emp)
    end = tech_data.find('\n', end+3)
```

```python
# Turn companies into data frame
df = pd.DataFrame.from_dict(company_dict, 'index')
df.rename(columns = {0 : 'Num_of_employees'}, inplace = True)
df.rename(columns = {1 : 'Zip_code'}, inplace = True)
```
